```python
#Task 7: Add “Mr.” and “Mrs.” to the first name column and drop the gender column.
```

```python
import pandas as pd    # importing pandas
path= "C:\\Users\\Mohini\\PY-Capstone\\demote.csv"  #reading the "demote" csv file
mr_mrs = pd.read_csv(path, index_col=0)
```

```python
# Adding "Mr." or "Mrs." before First Name
mr_mrs["First Name"] = mr_mrs.apply(lambda row: f"Mr. {row['First Name']}" if row["Gender"] == "Male" else f"Mrs. {row['First Name']}", axis=1)
```

```python
mr_mrs=mr_mrs.drop(columns=['Gender'])  #Dropping the "gender" column
```

```python
mr_mrs.to_csv("mr_mrs.csv")   # saving "mr_mrs" dataframe as csv file for further use
```

```python
#Task 8: Promote designation level by 1 for the employees whose age is more than 29 years using IF condition.
```

```python
path1= "C:\\Users\\Mohini\\PY-Capstone\\mr_mrs.csv"  #reading the "mr_mrs" csv file
promote = pd.read_csv(path1, index_col=0)
```

```python
#promoting employees designation level, whose age is more than 29 
promote["Designation Level"] = promote.apply(lambda row: row["Designation Level"]+1  if row["Age"] > 29 else row["Designation Level"], axis=1)
promote
```
