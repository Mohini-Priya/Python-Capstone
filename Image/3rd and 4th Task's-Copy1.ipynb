```python
#Task 3: Split the name column in the Employee dataframe into two new columns “First Name”, and “Last Name” and remove the older “name” column.
```

```python
import pandas as pd    # importing pandas
path= "C:\\Users\\Mohini\\PY-Capstone\\employee.csv"    #reading the "employee" csv file
df = pd.read_csv(path, index_col=0)
```

```python
df[['First Name', 'Last Name']] = df["Name"].str.split(' ', 1, expand=True) # Splitting the "Name" column into "First Name" and "Last Name"
df.drop(columns=['Name'], inplace=True)  # Dropping the "Name" column
```

```python
print(df.to_string(index=False))
```

```python
emp=df
emp
```

```python
emp.to_csv('emp.csv')#saving the dataframe(for which first name and last name columns were added and Name column was dropped) as csv file for further use
```

```python
#Task 4: Join all three dataframes in one single dataframe. Name it “Final”

path1= "C:\\Users\\Mohini\\PY-Capstone\\newproject.csv"              #reading the "newproject" csv file
path2= "C:\\Users\\Mohini\\PY-Capstone\\emp.csv"                     #reading the "emp" csv file
path3= "C:\\Users\\Mohini\\PY-Capstone\\senioritylevel.csv"          #reading the "senioritylevel" csv file
df1 = pd.read_csv(path1, index_col=0)
df2 = pd.read_csv(path2, index_col=0)
df3 = pd.read_csv(path3, index_col=0)
```

```python
final = df1.merge(df2, on="ID").merge(df3, on="ID")
```
