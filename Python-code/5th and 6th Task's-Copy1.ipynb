```python
#Task 5: Add a new bonus column in the Final dataframe. 
#Give a 5% bonus concerning project cost only to employees who have finished the projects.
```

```python
import pandas as pd    # importing pandas
path= "C:\\Users\\Mohini\\PY-Capstone\\final.csv"  #reading the "final" csv file
df = pd.read_csv(path, index_col=0)
```

```python
pd.options.display.float_format = '{:.0f}'.format  # Disable scientific notation

# Applying 5% bonus for employee's, whose project status is "Finished"
df["Bonus"] = df.apply(lambda row: row["Cost"] * 0.05 if row["Status"] == "Finished" else 0, axis=1)
Bonus=df             #storing the "df" dataframe in new variable
Bonus
```

```python
Bonus.to_csv("Bonus.csv")   #saving the "Bonus" dataframe as csv file for further use
```
